# Data Processing

After extracting data from Google BigQuery. This python script will be used to process data specifc to ENS

```python
# import to invoke required module
import pandas as pd
import numpy as np
from tabulate import tabulate
from prettytable import PrettyTable
# Dataframe display options
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 5000)
```

# Post ENS Data Preparation

```python
dfTrace = pd.read_csv('/Users/rparama/Downloads/Dataset/1_ETH_DataPrepration.csv')
dfTrace.head()
```

```python
# File metadata/Schema analysis


headers=PrettyTable(["Column Name","Record Count","Unique Count","Null Count","Total Count"])
for col in dfTrace.columns:
    headers.add_row([col,dfTrace[col].count(),dfTrace[col].nunique(),
                    dfTrace[col].isnull().sum(),dfTrace[col].count()+dfTrace[col].isnull().sum()])
print(headers)
```

```python
dfTrace = dfTrace[(dfTrace['block_timestamp_x'] == dfTrace['block_timestamp_y']) &
                  (dfTrace['block_number_x'] == dfTrace['block_number_y']) &
                  (dfTrace['block_hash_x'] == dfTrace['block_hash_y']) 
                 ]
```

```python
headers=PrettyTable(["S.No","Column Name","Record Count","Unique Count","Null Count","Total Count"])
i=1
for col in dfTrace.columns:
    headers.add_row([i,col,dfTrace[col].count(),dfTrace[col].nunique(),
                    dfTrace[col].isnull().sum(),dfTrace[col].count()+dfTrace[col].isnull().sum()])
    i+=1
    
print(headers)
```

```python

dfTrace.drop(['block_timestamp_y',
              'block_number_y',
              'block_hash_y'
             ], axis=1, inplace=True)
```

```python
dfTrace.columns = ['transaction_hash', 
                   'transaction_index',
                   'trace_from_address',
                   'trace_to_address',
                   'trace_value',
                   'trace_input',
                   'trace_output',
                   'trace_gas',
                   'trace_gas_used',
                   'subtraces',
                   'block_timestamp',
                   'block_number',
                   'block_hash',
                   'token_address',
                   'token_transfer_from_address',
                   'token_transfer_to_address',
                   'token_transfer_value',
                   'log_index'                
                  ]
```

# Contract

```python
dfContract = pd.read_csv('/Users/rparama/Downloads/Dataset/contracts_2019.csv')
dfContract.head()
```

```python

contract_from = pd.merge(dfETH2,dfContract, how = 'inner', left_on=['trace_from_address'],right_on=['address'])
```

```python
contract_from.head()
```

```python
contract_to = pd.merge(dfETH2,dfContract, how = 'inner', left_on=['trace_to_address'],right_on=['address'])
```

```python
contract_to.shape
```

```python
contract_to.head()
```

```python
contract_to.address.nunique()
```

```python
# Concat ENS and ENS history

dfENSConcat = pd.concat([contract_from,contract_to],axis=0)
dfENSConcat.head(10)
```

```python
headers=PrettyTable(["S.No","Column Name","Record Count","Unique Count","Null Count","Total Count"])
i=1
for col in dfENSConcat.columns:
    headers.add_row([i,col,dfENSConcat[col].count(),dfENSConcat[col].nunique(),
                    dfENSConcat[col].isnull().sum(),dfENSConcat[col].count()+dfENSConcat[col].isnull().sum()])
    i+=1
    
print(headers)
```

```python
dfENSConcat.drop(['block_timestamp_y',
            'block_number_y',
            'block_hash_y'
           ], axis=1, inplace=True)
```

```python
dfENSConcat.columns = ['transaction_hash',
                       'transaction_index',
                       'trace_from_address',
                       'trace_to_address',
                       'trace_value',
                       'trace_input',
                       'trace_output',
                       'trace_gas',
                       'trace_gas_used',
                       'subtraces',
                       'block_timestamp',
                       'block_number',
                       'block_hash',
                       'token_transfer_address',
                       'token_transfer_from_address',
                       'token_transfer_to_address',
                       'token_transfer_value',
                       'log_index',
                       'hash',
                       'nonce',
                       'transaction_from_address',
                       'transaction_to_address',
                       'transaction_value',
                       'transaction_gas',
                       'transaction_gas_price',
                       'transaction_input',
                       'receipt_cumulative_gas_used',
                       'receipt_gas_used',
                       'receipt_contract_address',
                       'receipt_root',
                       'receipt_status',
                       'token_address',
                       'token_symbol',
                       'token_name',
                       'token_decimals',
                       'token_total_supply',
                       'Contract_address',
                       'Contract_bytecode',
                       'Contract_function_sighashes',
                       'Contract_is_erc20',
                       'Contract_is_erc721',
                       'Contract_block_Date'
                  ]
```

```python
# removing null records
dfENSConcat = dfENSConcat[~dfENSConcat.token_symbol.isnull() & 
                          ~dfENSConcat.token_name.isnull() & 
                          ~dfENSConcat.token_decimals.isnull() &
                          ~dfENSConcat.transaction_to_address.isnull()
                         ]
```

```python
dfENSConcat["receipt_status"].fillna(0, inplace=True)
```

```python
dfENSConcat.drop(['receipt_contract_address',
            'receipt_root'
           ], axis=1, inplace=True)
```

```python
dfENSConcat.reset_index(inplace=True, drop=True)
```

```python
headers=PrettyTable(["S.No","Column Name","Record Count","Unique Count","Null Count","Total Count"])
i=1
for col in dfENSConcat.columns:
    headers.add_row([i,
                     col,
                     dfENSConcat[col].count(),
                     dfENSConcat[col].nunique(),
                     dfENSConcat[col].isnull().sum(),
                     dfENSConcat[col].count()+dfENSConcat[col].isnull().sum()])
    i+=1
    
print(headers)
```
