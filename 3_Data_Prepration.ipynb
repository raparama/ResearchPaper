```python

dfETH = pd.merge(trace,token_transfer, how = 'inner', left_on=['transaction_hash'],right_on=['transaction_hash'])
dfETH.head(10)
```

```python
dfETH.shape
```

```python
dfETH["token_address"].count(), dfETH["token_address"].nunique()
```

```python
# File metadata/Schema analysis


print("| Column Name","Record Count","Unique Count","Null Count","Total Count")
for col in dfETH.columns:
    print(([col.ljust(15),dfETH[col].count(),dfETH[col].nunique(),
                            dfETH[col].isnull().sum(),dfETH[col].count()+dfETH[col].isnull().sum()]))
```

```python

dfETH = dfETH[(dfETH["trace_type"] == 'call') &
              (dfETH["status"] == 1) & 
              (dfETH["call_type"] == 'call')].reset_index().drop('index',axis=1)
```
